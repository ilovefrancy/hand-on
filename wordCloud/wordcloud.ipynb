```python
from urllib import request
target_path = 'https://sh.lianjia.com/ershoufang/'
response = request.urlopen(target_path)
html_data = response.read().decode('utf-8')
```

```python
print(html_data)
```

```python
from bs4 import BeautifulSoup as bs
```

```python
soup = bs(html_data, 'html.parser')    
house = soup.find_all('ul', class_='sellListContent')
```

```python
print(house[0])
```

```python
house_list = house[0].find_all('li', class_='clear')
```

```python
print(house_list[3])
```

```python
house2_list = []
for item in house_list:        
        house_dict = {}        
        house_dict['id'] = item['data-housecode']      
        for tag_img_item in item.find_all('img'):            
            house_dict['name'] = tag_img_item['alt']            
            house_list.append(house_dict)
```
